```python
# Group by age and calculate the sum of the target
age_target_sum = df.groupby('age')['target'].sum().reset_index()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(age_target_sum['age'], age_target_sum['target'], marker='o', linestyle='-')
plt.title("Sum of Target by Age")
plt.xlabel("Age")
plt.ylabel("Sum of Target")
plt.grid(True)
plt.show()
```
